```python
from scipy.stats import expon
```

```python
import fincorrnets as fnx
import networkx as nx
```

```python
import numpy as np
import math
import pandas as pd
```

```python
α = 0.3
```

```python
x = np.linspace(0,4,100)
```

```python
1 - np.exp(-α)
```

```python
f = lambda x, α: expon.pdf(x, scale=1/α) / (1-np.exp(-α))
cdf = lambda x, α: expon.cdf(x, scale=1/α) / (1-np.exp(-α))
```

```python
cdf(1, α)
```

```python
np.sum(f(np.linspace(0,1,100), α))
```

```python
from matplotlib import pyplot as plt
```

```python
from bokeh.palettes import inferno
```

```python
%matplotlib inline
plt.figure(figsize=(7,3))
ax = plt.axes()
ax.hold(True)
xdata = np.linspace(0,1,100)
nlines = 8
colors = inferno(nlines)
for i,α in enumerate(np.linspace(0.01,5.,nlines)):
    ydata = f(xdata, α)
    plt.plot(xdata, ydata, axes=ax, color=colors[i])
```

```python
%matplotlib inline
plt.figure(figsize=(7,3))
ax = plt.axes()
ax.hold(True)
xdata = np.linspace(0,1,100)
nlines = 8
colors = inferno(nlines)
for i,α in enumerate(np.linspace(0.01,40.,nlines)):
    ydata = cdf(xdata, α)
    plt.plot(xdata, ydata, axes=ax, color=colors[i])
```
